## ガウスカーネルで確率密度算出

```python
# scipyモジュールをインポート
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# データのサンプル（正規分布から生成）
data = [0.5, 1.2, 1.8, 2.3, 2.7, 3.1]

# ガウスカーネルを用いたKDEを実行
kde = stats.gaussian_kde(data, bw_method=0.5)

# x軸の値の範囲
x = np.linspace(0, 4, 100)

# x軸の値に対応する確率密度を計算
y = kde(x)

# 結果をプロット
plt.plot(x, y)
plt.show()
```

## 線形カーネルで確率密度算出

```python
import numpy as np
import matplotlib.pyplot as plt
# scikit-learnモジュールをインポート
from sklearn.neighbors import KernelDensity

kernel = int(input("カーネル関数選択(1:ガウス,2:線形"))

# データのサンプル（正規分布から生成）
data = np.random.normal(0, 1, 100)

if kernel == 1:
    title = "Gaussian"
    kde = KernelDensity(kernel="gaussian", bandwidth=0.5).fit(data[:, None])
elif kernel == 2:
    # 線形カーネルを用いたKDEを実行
    # バンド幅を0.2に指定
    title = "Linear"
    kde = KernelDensity(kernel="linear", bandwidth=0.2).fit(data[:, None])

# x軸の値の範囲
x = np.linspace(-5, 5, 100)[:, None]

# x軸の値に対応する確率密度を計算
dens = kde.score_samples(x)

# 結果をプロット
plt.title(title)
plt.plot(x, np.exp(dens))
plt.show()
```
